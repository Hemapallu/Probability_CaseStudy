## Marginal Probabilities(Single Column)

1.What is the probability of getting rain tomorrow?

```python
sum(weather['Outlook']=='Rain')/len(weather)
```

```python
weather['Outlook'].value_counts(normalize=True).loc['Rain']
```

2.What is the probability of playing cricket tomorrow?

```python
sum(weather['Play']=='Yes')/len(weather)
```

```python
weather['Play'].value_counts(normalize=True).loc['Yes']
```

3.what is the probability that temperature is not cool tomorrow?

```python
sum(weather['Temperature']!='Cool')/len(weather)
```

```python
1-weather['Temperature'].value_counts(normalize=True).loc['Cool']
```

### Joint Probability

1.What is the probability of playing cricket and temperature is hot?

```python
sum((weather['Play']=='Yes')&(weather['Temperature']=='Hot'))/len(weather)
```

```python
pd.crosstab(index=weather['Play'],columns=weather['Temperature'],normalize=True).loc['Yes','Hot']
```

```python
pd.crosstab(index=weather['Play'],columns=weather['Temperature'],normalize=True).loc['Yes']['Hot']
```

2.what is the probability of humidity is normal and temperature is cool?

```python
pd.crosstab(index=weather['Humidity'],columns=weather['Temperature'],normalize=True).loc['Normal']['Cool']
```

### Conditional Probability

1.What is the probability of playing cricket if it is raining?

```python
pd.crosstab(index=weather['Play'],columns=weather['Outlook'],normalize='columns').loc['Yes','Rain']
```

2.What is the probability of not playing cricket when humidity is high?

```python
pd.crosstab(index=weather['Humidity'],columns=weather['Play'],normalize='index').loc['High','No']
```
